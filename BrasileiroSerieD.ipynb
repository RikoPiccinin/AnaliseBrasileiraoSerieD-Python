```python
df_tabela = df_tabela.rename(columns={'PTSPontos' : 'Pontos'})
```

```python
df_tabela = df_tabela.rename(columns={'VVitórias' : 'Vitorias', 'JJogos': 'Jogos', 'EEmpates': 'Empates', 'DDerrotas': 'Derrotas'})
```

```python
df_tabela = df_tabela.rename(columns={'GPGols Prós': 'Gols_Pro', 'GPGols Contras': 'Gols_Contras' , 'SGSaldos de Gols' : 'Saldo_Gols', 'CACartões Amarelos' : 'Cartoes_Amarelos', 'CVCartões Vermelhos' : 'Cartoes_Vermelhos'})
```

```python
df_tabela
```

```python
df_tabela = df_tabela.rename(columns={'Classificação' : 'Times'})
```

```python
df_tabela = df_tabela.drop(columns=['Unnamed: 0'])
```

```python
#LIMPANDO A COLUNA TIME, QUE APRESENTA NUMEROS E CARACTERES JUNTO
```

```python
df_tabela['Times'] = df_tabela['Times'].str.replace(r'[^A-Za-zÀ-ÿ\s]', '', regex=True)
```

```python
gols_por_time = df_tabela.groupby('Times')['Saldo_Gols'].sum().sort_values(ascending=False)
print(gols_por_time)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
gols_por_time.plot(kind='bar', color='skyblue')
plt.title('Gols por Time')
plt.xlabel('Time')
plt.ylabel('Gols')
```

```python
gols_por_time = df_tabela.groupby('Times')['Saldo_Gols'].sum().sort_values(ascending=False)


plt.figure(figsize=(16, 8))
gols_por_time.plot(kind='bar')

plt.title('Gols por Time')
plt.xlabel('Times')
plt.ylabel('Total de Gols')
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```

```python
# Criar gráfico
plt.figure(figsize=(16, 8))
ax = gols_por_time.plot(kind='bar')

plt.title('Gols por Time')
plt.xlabel('Times')
plt.ylabel('Total de Gols')
plt.xticks(rotation=75)

# Adicionar rótulos nas barras
for container in ax.containers:
    ax.bar_label(container, label_type='edge')  # Mostra o valor no topo

plt.tight_layout()
plt.show()
```
